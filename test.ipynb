```python
from sentence_transformers import SentenceTransformer, util

# Load model
model = SentenceTransformer("intfloat/multilingual-e5-large")

# 🔍 Query và documents (đã thêm prefix theo đúng format của E5)
query = "query: phím chức năng"
documents = [
    "passage: Phím nguồn (1): Ấn giữ để bật/tắt thiết bị.",
    "passage: Cổng kết nối HDMI để xuất hình ảnh.",
    "passage: Thiết bị nội khí quản có camera tích hợp.",
    "passage: Hướng dẫn sử dụng nhanh bộ đặt nội khí quản."
]

# Encode
query_embedding = model.encode(query, convert_to_tensor=True)
doc_embeddings = model.encode(documents, convert_to_tensor=True)

# Tính cosine similarity
cos_scores = util.cos_sim(query_embedding, doc_embeddings)

# In kết quả
for i, score in enumerate(cos_scores[0]):
    print(f"Document {i+1}: Score = {score:.4f} - {documents[i][9:]}")
```
