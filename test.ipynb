```python
from sentence_transformers import SentenceTransformer, util

# Load model
model = SentenceTransformer("intfloat/multilingual-e5-large")

# üîç Query v√† documents (ƒë√£ th√™m prefix theo ƒë√∫ng format c·ªßa E5)
query = "query: ph√≠m ch·ª©c nƒÉng"
documents = [
    "passage: Ph√≠m ngu·ªìn (1): ·∫§n gi·ªØ ƒë·ªÉ b·∫≠t/t·∫Øt thi·∫øt b·ªã.",
    "passage: C·ªïng k·∫øt n·ªëi HDMI ƒë·ªÉ xu·∫•t h√¨nh ·∫£nh.",
    "passage: Thi·∫øt b·ªã n·ªôi kh√≠ qu·∫£n c√≥ camera t√≠ch h·ª£p.",
    "passage: H∆∞·ªõng d·∫´n s·ª≠ d·ª•ng nhanh b·ªô ƒë·∫∑t n·ªôi kh√≠ qu·∫£n."
]

# Encode
query_embedding = model.encode(query, convert_to_tensor=True)
doc_embeddings = model.encode(documents, convert_to_tensor=True)

# T√≠nh cosine similarity
cos_scores = util.cos_sim(query_embedding, doc_embeddings)

# In k·∫øt qu·∫£
for i, score in enumerate(cos_scores[0]):
    print(f"Document {i+1}: Score = {score:.4f} - {documents[i][9:]}")
```
